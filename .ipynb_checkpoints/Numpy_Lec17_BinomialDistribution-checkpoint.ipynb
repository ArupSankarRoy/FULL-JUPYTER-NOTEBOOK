```python
# BINOMIAL DIST - discrete dist - binary output
# parameter :: n(Number of trials like Coin toss) , p(Probability) , size(Shape-return array)
```

```python
from numpy import random
import seaborn as sns
import matplotlib.pyplot as plt

sns.distplot(random.binomial(n= 10 , p = 0.5 , size =1000) , hist = False , kde=True)
plt.show()
```

```python
from numpy import random
import seaborn as sns
import matplotlib.pyplot as plt

sns.distplot(random.normal(loc = 50 , scale = 5 , size = 1000), hist = False , label = "Normal")
sns.distplot(random.binomial(n = 100 , p = 0.5 , size = 500) , hist = False , label = "Binomial")
plt.show()
```
