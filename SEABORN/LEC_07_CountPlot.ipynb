```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

# Calculate the total number of observations
total = float(len(tips))

# Create a countplot
ax = sns.countplot(x="smoker",hue = "sex" , data=tips)

# Convert the y-axis values to percentages
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:.1f}%'.format((height/total)*100), # height/total calculates the proportion of the total number of observations represented by the current patch,
            ha="center")
 

plt.show()

# NOTE ::
# In the context of Seaborn's countplot, "patches" refers to the individual rectangular
# bars that represent the count of observations in each category. These bars are called "patches"
# in Matplotlib, which is the underlying visualization library used by Seaborn.
# p represents each individual patch (bar) in the countplot, and we retrieve the height of each patch using the get_height() method.
# We then use this height value to add a text label above the patch, indicating the percentage of total observations it represents.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

# Calculate the total number of observations
total = float(len(tips))

# Create a countplot
ax = sns.barplot(x="smoker", y = "size", hue = "sex" , data=tips)

# Convert the y-axis values to percentages
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height + 0.25,
            '{:.1f}%'.format((height/total)*100), # height/total calculates the proportion of the total number of observations represented by the current patch,
            ha="center")
 

plt.show()
```
