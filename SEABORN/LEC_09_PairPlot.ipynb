
# PAIR PLOT

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset("tips")
df
```

```python
sns.pairplot(df) # NOTE :Plots are made btw the numerical columns 
plt.show()
```

```python
sns.pairplot(df , hue = "sex") 
plt.show()
```

## PLOTTING PAIR PLOT ONLY FOR SPECIFIC COLUMNS

```python
sns.pairplot(df , vars = ["total_bill" , "tip"] , hue = "sex" ,
             hue_order = ["Female" , "Male"] ,) 
plt.show()
```

```python
sns.pairplot(df , hue = "sex" ,
             hue_order = ["Female" , "Male"] , y_vars = ["total_bill" , "tip"]) 
plt.show()
```

```python
sns.pairplot(df , hue = "sex" ,
             hue_order = ["Female" , "Male"] , kind = "reg" ,
             diag_kind = "hist" , markers = ["*" , "<"])

# parameter kind = "scatter","kde","reg","hist"..etc
# diag_kind is basically the DIAGONAL PLOTTING
plt.show()
```
