# SCATTER PLOT

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset("penguins").head(20)
df
```

```python
m = {"Male" : "o" , "Female" : "*"}
sns.scatterplot(x = "bill_length_mm", y="bill_depth_mm",hue = "sex" ,
                style = "sex", size = "sex",sizes = (100,30),
                palette = "gist_ncar",markers = m, data=df)
plt.show()
```
