# STRIP PLOT

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df
```

```python
# The jitter argument in Seaborn's stripplot() function is used to add a small amount of random noise
# to the data points along the categorical axis. This is done to avoid overlapping of points when there 
# are multiple data points with the same value along the categorical axis.
# By default, jitter is set to True, which adds a small amount of random noise to the data points along the 
# categorical axis. This causes the points to spread out a bit, making it easier to see individual data points.

sns.stripplot(x = "day" , y = "total_bill" ,hue = "sex",
              data = df , palette = "rocket_r" , linewidth = 1,
              edgecolor = "y" , jitter = 3 )
plt.show()
```
