# Bar Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("penguins")
df
```

```python
sns.barplot(x = df.island , y = df.bill_length_mm ,hue = df.sex , palette = "icefire" , saturation = 0.3 , errcolor = "g" ,
            errwidth = 4 , capsize = 0.3)
# OR , sns.barplot(x = "island" , y = "bill_length_mm" , data = df)

#     __
#     |
#    _|_

plt.show()
```

```python
sns.set(style = "darkgrid")

order_1  = ["Dream" , "Torgersen", "Biscoe"]
sns.barplot(x = df.island , y = df.bill_length_mm ,hue = df.sex , hue_order = ["Female" , "Male"]  , order = order_1 , ci = 90 , n_boot = 11)
plt.show()
# here "ci" argument adjust the error bars(Black bars on top of the seaborn bars) height(range : 0-100)
#  dodge = False seperate the Male and Female Error
```

### "ORIENT" PARAMETER WORKS ONLY ON NUMERICAL VALUE

```python

sns.barplot(x = df.bill_length_mm , y = df.	bill_depth_mm ,hue = df.sex , hue_order = ["Female" , "Male"] , orient = "h")
plt.show()
```
