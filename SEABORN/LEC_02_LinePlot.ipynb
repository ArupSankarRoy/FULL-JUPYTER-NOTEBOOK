# Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var = [1,2,3,4,5,6,7,8]
var_1 = [5,9,7,8,6,7,2,5]

df = pd.DataFrame({"var" : var , "var_1" : var_1})

sns.lineplot(x = "var" , y = "var_1" , data = df)

plt.show()
```

```python
plt.plot(var , var_1)
plt.show()
```

## USING SEABORN PLOTTING MATPLOTLIB GRAPH

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var = [1,2,3,4,5,6,7,8]
var_1 = [5,9,7,8,6,7,2,5]

# WITHOUT- DATA FRAME
sns.lineplot(x = var , y = var_1)

plt.show()
```

### LOADING CSV FILE WITHOUT DOWNLOADING IT

```python
y_1 = sns.load_dataset("penguins").head(20) # NOTE REMOVE .csv EXTENSION
y_1
```

```python
sns.lineplot(x = "bill_depth_mm" , y = "flipper_length_mm" ,hue = "sex" ,
             style = "sex",palette = "Accent" , markers = ["o" , ">"],legend = False, data = y_1) # This graph is made with respecct to "sex"
# (dashes = False) used to remove dotted lines.

plt.grid()
plt.title("Python")
plt.show()
```
