# CAT PLOT

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df
```

```python
sns.catplot(x="tip", y="size",hue = "sex", data=df)
plt.show()
```

```python
sns.catplot(x="day", y="size",hue = "sex",kind = "point", data=df)
plt.show()
```
