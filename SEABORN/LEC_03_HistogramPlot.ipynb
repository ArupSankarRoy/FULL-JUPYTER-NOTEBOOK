# HISTOGRAM PLOT

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset("penguins")
df
```

```python
sns.displot(df["bill_length_mm"] , bins = [30 , 35 , 40 , 45 , 50 , 55 , 60 , 65] , kde = True , rug = True ,color ="r") # bins used for Organise the Data bars and fixing the range of each bar.

# We also use log_scale = True to represent graph data as log format
plt.show()
```
