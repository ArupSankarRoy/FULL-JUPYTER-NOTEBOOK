## VIOLIN PLOT

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset("tips")
df
```

```python
sns.violinplot(x = df.day , y = df.total_bill , hue = df.time , linewidth = 2)
plt.show()
```

```python
sns.violinplot(x = df.time , y = df.total_bill , order = ["Dinner" , "Lunch"])
plt.show()
```

```python
sns.violinplot(x = df.day , y = df.total_bill , hue = df.sex , split = True , scale = "count")
plt.show()

# Parameters of scale = "count" , "area" , "width"...etc
```

```python
sns.violinplot(x = df.day , y = df.total_bill , hue = df.sex ,inner = "point")

# Parameters of inner = "quart","box","quartile","point","stick",None....etc
plt.show()
```

```python
sns.violinplot(x = df["total_bill"])
plt.title("SINGLE VIOLIN PLOT" , color = "r" , size = 25)
plt.show()
```
