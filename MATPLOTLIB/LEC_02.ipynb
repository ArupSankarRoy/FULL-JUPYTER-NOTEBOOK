```python
import matplotlib.pyplot as plt
x = [1 , 2 , 3 , 4 , 5]
y = [6 , 7 , 8 , 9 , 10]

plt.figure(figsize = (8 , 4))

plt.subplot(1 , 2 , 1)
plt.plot(x , y , color = "r")
plt.xlabel("x axis -->")
plt.ylabel("y axis -->")

plt.subplot(1 , 2 , 2)
c = ["r" , "y" , "g" , "b" , "m"]
plt.bar(x , y , color = c)
plt.xlabel("x axis -->")
plt.ylabel("y axis -->")

plt.show()
```
