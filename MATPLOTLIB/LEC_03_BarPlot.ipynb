# BAR PLOT

```python
import matplotlib.pyplot as plt

x = ["Python" , "Java" ,"C++" , "C#"]
y = [85 , 80 , 70 , 75]
c = ["b" , "r" , "y" , "m"]

plt.title("Bar plot")
# plt.bar(x , y , width = 0.4 ,color=c , edgecolor="g", linewidth=5 , linestyle=":" , alpha=0.7 , align="edge") # SEE THE "Python" , "Java" ,"C++" , "C#" ALIGNMENT IN BAR

# THE VALUE OF ALPHA IS BETWEEN 0 TO 1 AND IT'S A FLOAT VALUE AND IT MAKES THE BAR COLOR DULL.
# EDGE COLOR IS THE BOUNDARY COLOR OF BARS.
# LINE WIDTH MODIFY THE BAR'S BORDER WIDTH ACCORDING TO IT'S VALUE.
# plt.xlabel("language popularity" , fontsize = 10)

plt.bar(x , y , width = 0.4 ,color=c ,align="edge" , label="Popularity Bar")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ["Python", "Java", "C++", "C#"]
y = [85, 80, 60, 62]
z = [20, 30, 40, 50]

plt.xlabel("Language", fontsize=15)
plt.ylabel("No", fontsize=15)

colors = ["b", "r"]
labels = ["Popularity 1", "Popularity 2"]
plt.plot(x, y, linewidth=2, c=colors[0], label=labels[0])
plt.plot(x, z, linewidth=2, c=colors[1], label=labels[1])

plt.legend()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["Python", "Java", "C++", "C#"]
y = [85, 80, 60, 62]
z = [20, 30, 40, 50]

width = 0.2
p = np.arange(len(x))
p1 = [j + width for j in p]  # use a comma to separate the expressions

plt.xlabel("Language", fontsize=15, color="m")
plt.ylabel("No", fontsize=15, color="m")

colors = ["b", "r"]
labels = ["Popularity 1", "Popularity 2"]
plt.bar(p1, y, width, color=colors[0], label=labels[0])
plt.bar(p1, z, width, color=colors[1], label=labels[1])

# set the x-tick positions and labels
plt.xticks(p1, x)

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["Python", "Java", "C++", "C#"]
y = [85, 80, 60, 62]
z = [20, 30, 40, 50]

width = 0.2
p = np.arange(len(x)) # [0 1 2 3]

p1 = p - width/2  # adjust the x-coordinates for the first set of bars
p2 = p + width/2  # adjust the x-coordinates for the second set of bars and it means p2 is one step ahead of p1

# distance between any set of bars is =(p2 - p1)= (p+width/2) - (p-width/2) = here it's 0.2


plt.xlabel("Language", fontsize=15, color="m")
plt.ylabel("No", fontsize=15, color="m")

colors = ["b", "r"]
labels = ["Popularity 1", "Popularity 2"]
plt.bar(p1, y, width, color=colors[0], label=labels[0])
plt.bar(p2, z, width, color=colors[1], label=labels[1])

# set the x-tick positions and labels
plt.xticks(p, x)

plt.legend()
plt.show()
```
