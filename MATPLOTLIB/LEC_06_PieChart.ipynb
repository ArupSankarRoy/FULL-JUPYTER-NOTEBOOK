# PIE PLOT

```python
import matplotlib.pyplot as plt

x = [10 , 20 , 30 , 40] # for 100%
y = ["C" , "C++" , "Java" , "Python"]
ex = [0.3 , 0.0 , 0.2 , 0.0]
c = ["m" , "y" , "r" , "b"]

#plt.pie(x , labels = y , explode = ex , colors=c , autopct = "%0.2f%%") # f represents it's a float value , 0.2 means in piechart
                                                                         # there will be 2 index after point
                                                                         # explode means the slice will be slightly move backward from the chart based on it's value
plt.pie(x , labels = y , explode = ex , colors=c , autopct = "%0.2f%%" , shadow = True , radius = 1.5 , labeldistance = 0.8 , startangle = 90)
# labeldistance means "C" , "C++" , "Java" , "Python" etc label's distance from the chart.
# normally any circle is start from angle 0. here use of startangle = 90, i state that now it's start from angle 90
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40] # for 100%
y = ["C", "C++", "Java", "Python"]
ex = [0.3, 0.0, 0.2, 0.0]
c = ["m", "y", "r", "b"]

textprops = {"fontsize": 13, "color": "w"}

plt.pie(x, labels=y, explode=ex, colors=c, autopct="%0.2f%%", shadow=True, radius=1.3,
        labeldistance=0.7, startangle=90, textprops=textprops , counterclock = False , wedgeprops={"linewidth" :5 , "edgecolor" : "g"}) # default counter is True ,It's related to
                                                                                                                    # chart's rotation-> clock and anti-clock
                                                                                                   # linewidth related to shadow's width                 

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40] # for 100%
y = ["C", "C++", "Java", "Python"]
ex = [0.3, 0.0, 0.2, 0.0]
c = ["m", "y", "r", "b"]

textprops = {"fontsize": 13, "color": "w"}

plt.pie(x, labels=y, explode=ex, colors=c, autopct="%0.2f%%", shadow=True, radius=1,
        labeldistance=0.7, startangle=90, textprops=textprops ,
        counterclock = False , wedgeprops={"linewidth" :5 } , center =(-3,0) , rotatelabels = False)  # labels like java , c etc are rotated               

plt.title("PIE CHART")
plt.legend(loc=5)
plt.show()
```

# DONUT CHART

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40]
y = ["C", "C++", "Java", "Python"]
x1 = [40 , 30 , 20 , 10]

c = ["r" , "b" , "g" , "y"]

textprops = {"fontsize": 13, "color": "w"}

plt.pie(x, labels=y , radius = 1.2)     
cr = plt.Circle(xy = (0,0) , radius = 1 , facecolor = "w")
plt.gca().add_artist(cr)

plt.title("PIE CHART")
plt.show()
```
