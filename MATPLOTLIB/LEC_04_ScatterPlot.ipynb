# SCATTER PLOT


```python
import matplotlib.pyplot as plt

x = [1 , 2 , 3 , 4 , 5 , 7]
y = [2 , 3 , 1 , 4 , 5 , 3]

color = ["r" , "b" , "g" , "m" , "y" , "r"] 
size = [500 , 200 , 300 , 700 , 100 , 400]

# plt.scatter(x, y , c = color , s=size , alpha =0.5 , marker = "*" , edgecolor="r" , linewidth=4)

# HERE EACH DOT SIZE IS DIFFERENT
# EDGECOLOR MEANS EACH DOT'S BOUNDARY COLOR
# LINEWIDTH IS BASICALLY THE BORDER WIDTH OF EACH DOT i.e '*'

colorMAP = [10 , 49 , 30 , 29 , 56  , 60] # range : 0-100
plt.scatter(x, y, s=size, c=colorMAP, cmap="BrBG") # cmap is a build in variable which contains a color-set where each color is define as a number which is in 
                                                   # a range between 0 - 100
t = plt.colorbar()
t.set_label("Color bar" , fontsize = 12 , color="b")

plt.title("SCATTER PLOT" , color = "r" , fontsize = 15)
plt.xlabel("Day ->" , fontsize=15 , color="m")
plt.ylabel("No ->" , fontsize=15 , color="m")
plt.show()
```

# 2 SCATTER PLOT IN ONE GRAPH

```python
import matplotlib.pyplot as plt

x = [1 , 2 , 3 , 4 , 5 , 7]
y = [2 , 3 , 1 , 4 , 5 , 3]

z = [9 , 8 , 7 , 6  , 10 , 17]

color = ["r" , "b" , "g" , "m" , "y" , "r"] 
size = [500 , 200 , 300 , 700 , 100 , 400]

colorMAP = [10 , 49 , 30 , 29 , 56  , 60] # range : 0-100
plt.scatter(x, y, s=size, c=colorMAP, cmap="BrBG")
plt.scatter(x, z, s=size, c="b" , marker = "*" , edgecolor = "r")
t = plt.colorbar()
t.set_label("Color bar" , fontsize = 12 , color="b")

plt.title("SCATTER PLOT" , color = "r" , fontsize = 15)
plt.xlabel("Day ->" , fontsize=15 , color="m")
plt.ylabel("No ->" , fontsize=15 , color="m")
plt.show()
```
