```python
import matplotlib.pyplot as plt
import numpy as np

number = [19, 40, 24, 25, 30, 33, 33, 50, 29, 51, 35, 44, 21, 10, 25, 26, 40, 10, 53, 12, 59, 35, 16, 54,
 24, 33, 42, 39, 29, 49, 46, 48, 22, 30, 10, 25, 34, 28, 19, 55, 57, 55, 39, 14, 10, 25, 55, 19,
 18, 42]
l = [10 , 20 , 30 , 40 , 50 , 60]
plt.hist(number , color="b" , edgecolor="r" , bins=l , orientation="horizontal" , rwidth = 0.8) 

# The default value of rwidth is 0.9 which means that the bars will have a width of 90% of the width of the bin.
# By changing the value of rwidth, we can adjust the spacing between the bars in the histogram plot.
# In the given code, rwidth=0.8 is used which means that the width of the bars will be 80% of the width of the bin.
# First it captures 100% width of the bin .Now it reduce to 80 % ,20% width is deducted from each of the bars.
plt.title("Histogram plot" , color="r" , fontsize = 15)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

number = [19, 40, 24, 25, 30, 33, 33, 50, 29, 51, 35, 44, 21, 10, 25, 26, 40, 10, 53, 12, 59, 35, 16, 54,
 24, 33, 42, 39, 29, 49, 46, 48, 22, 30, 10, 25, 34, 28, 19, 55, 57, 55, 39, 14, 10, 25, 55, 19,
 18, 42]
l = [10 , 20 , 30 , 40 , 50 , 60]
plt.hist(number , color="b" , edgecolor="r" , bins=l ,log = True , label = "Python") 
plt.xlabel("LOG TABLE" , fontsize = 15 , color="m")
plt.title("Histogram plot" , color="r" , fontsize = 15)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

number = [19, 40, 24, 25, 30, 33, 33, 50, 29, 51, 35, 44, 21, 10, 25, 26, 40, 10, 53, 12, 59, 35, 16, 54,
 24, 33, 42, 39, 29, 49, 46, 48, 22, 30, 10, 25, 34, 28, 19, 55, 57, 55, 39, 14, 10, 25, 55, 19,
 18, 42]
l = [10 , 20 , 30 , 40 , 50 , 60]

plt.hist(number , edgecolor="r" , bins=l ,log = True , label = "Python") 
plt.xlabel("LOG TABLE" , fontsize = 15 , color="m")
plt.title("Histogram plot" , color="r" , fontsize = 15)
plt.axvline(40 , color="g", label="Value 1")
plt.axvline(50 , color="r", label="Value 2")
plt.grid()
plt.legend()
plt.show()
```
