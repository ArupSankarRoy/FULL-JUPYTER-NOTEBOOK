```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
```

```python
X,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1,noise=80)
```

```python
y
```

```python
plt.scatter(X , y)
plt.show()
```

```python
# Lets apply OLS
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
print('LR m is {} and LR b is {} '.format(reg.coef_ , reg.intercept_))
```

```python
plt.scatter(X , y)
plt.plot(X , reg.predict(X) , color = 'r' , label = 'OLS')
plt.legend()
plt.show()
```

```python
# THIS IS A RANDOM LINE FROM THIS LINE I MOVE CLOSE TO ACTUAL REGRESSION LINE KEEPING m CONSTANT
# Lets apply Gradient Descent assuming slope is constant m = 58.40831128
# and let's assume the starting value for intercept b = 0
y_pred = ((58.40831128 * X) + 100).reshape(4)
```

```python
plt.scatter(X , y)
plt.plot(X , reg.predict(X) , color = 'r' , label = 'OLS')
plt.plot(X , y_pred , color = 'b' , label = 'b = 0')
plt.legend()
plt.show()
```

```python
m = 58.40831128
b = 0

loss_slope = -2 * np.sum(y - m * X.ravel() - b)
loss_slope
```

```python
# Lets take learning rate = 0.1
lr = 0.1
step_size = lr * loss_slope
step_size
```

```python
# Calculating the new intercept
b = b - step_size
b
```

```python
y_pred1 = ((m * X) + b).reshape(4)
plt.scatter(X , y)
plt.plot(X , reg.predict(X) , color = 'r' , label = 'OLS')
plt.plot(X , y_pred , color = 'b' , label = 'b = 0')
plt.plot(X , y_pred1 , color = 'g' , label = 'b = 12.450041')
plt.legend()
plt.show()
```

```python
m = 58.40831128
b = -1000
epochs = 100

for i in range(epochs):
    
    loss_slope = -2 * np.sum(y - m * X.ravel() - b)
    step_size = 0.01 * loss_slope
    b = b - step_size
    
    y_pred = (m * X + b)
    plt.plot(X , y_pred)

plt.scatter(X , y)
plt.plot(X , reg.predict(X) , color = 'k' , label = 'OLS')
plt.legend()    
plt.show()


    
```
