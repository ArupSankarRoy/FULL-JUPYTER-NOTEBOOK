```python
import matplotlib.pyplot as plt

x_i = [0, 1, 2, 3, 4]
y1 = [1, 2, 3, 2, 1]
y2 = [2, 3, 4, 3, 2]

fig, ax = plt.subplots()

line1, = ax.plot(x_i, y1, 'r-', linewidth=2)
line2, = ax.plot(x_i, y2, 'b--', linewidth=2)

# Customize each line separately
line1.set_label('Line 1')
line2.set_label('Line 2')

ax.legend()
plt.show()
```
