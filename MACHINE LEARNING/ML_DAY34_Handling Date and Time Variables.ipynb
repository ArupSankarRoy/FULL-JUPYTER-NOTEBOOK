# Working with Dates

```python
# Converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
# Extracting year
date['date_year'] = date['date'].dt.year
date
```

```python
# Extracting month
date['date_month'] = date['date'].dt.month
date
```

```python
date['date_month_name'] = date['date'].dt.month_name()
date
```

```python
# Extract Days
date['date_day'] = date['date'].dt.day
# Extract Day naame
date['date_day_name'] = date['date'].dt.day_name()
date
```

```python
# day of week

date['date_dow'] = date['date'].dt.dayofweek
date
```

```python
# is weekend ?
date['date_is_weekend'] = np.where(date['date_day_name'].isin(['Sunday' , 'Saturday']) , 1 , 0)
date
```

```python
date.drop(['product_id' , 'city_id' , 'orders'] , axis = 1).head()
```

```python
date
```

```python
# Extract week of the year

date['date_week'] = date['date'].dt.isocalendar().week
date
```

```python
# Extract quarter 
date['date_quarter'] = date['date'].dt.quarter
date
```

# Extract time elapsed between dates

```python
import datetime

today = datetime.datetime.today()
today
```

```python
(today - date['date']).dt.days
```

```python
# Months passed
np.round((today - date['date'])/np.timedelta64(1,'M') , 0)
```
