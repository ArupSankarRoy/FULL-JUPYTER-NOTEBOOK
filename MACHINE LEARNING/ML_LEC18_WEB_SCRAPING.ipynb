```python
company = soup.find_all('div' , class_ = 'resultbox_textbox')
```

```python
names = []
rating = []
reviews = []
ctype = []
hq = []
old = []
freq = 0

for i in company:
    h2_element = i.find('h2')  # Assuming `company` is a collection of HTML elements
    if h2_element is not None:
        names.append(h2_element.text.strip())
    else:
        names.append("")  # Append an empty string if `<h2>` element is not found
    
    rating_element = i.find('div', class_='resultbox_totalrate mr-6 font14 fw700 colorFFF')
    if rating_element is not None:
        rating.append(rating_element.text.strip())
    else:
        rating.append("")  # Append an empty string if `<div>` element is not found
    reviews_element = i.find('div' , class_='resultbox_countrate ml-12 mr-12 font14 fw400 color777')
    if reviews_element is not None:
        reviews.append(reviews_element.text.strip())
    else:
        reviews.append("")
        
    ctype_element = company[freq].find('span', class_='amenities_tabs font12 fw500 color777')
    if ctype_element is not None and ctype_element.text is not None:
        ctype.append(ctype_element.text.strip())
        freq = freq + 1
    else:
        ctype.append("")
        
df = pd.DataFrame({'names': names , 'ratings': rating , 'reviews': reviews , 'company type': ctype})
df

    
```
