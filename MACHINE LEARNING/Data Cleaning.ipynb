```python
df.drop(['Serial No'] , axis = 1 , inplace = True)
```

```python
ls = ['Serial No.1', 'Roll No']
ls_before_roll = []
ls_after_roll = []
ls_before_SerialNo = []
ls_after_SerialNo = []

regex = re.compile('\n')

for col in ls:
    for val in df[col].astype(str).values:
        if regex.search(val):
            a, b = val.split('\n')
            if col == 'Roll No':
                ls_before_roll.append(a)
                ls_after_roll.append(b)
            elif col == 'Serial No.1':
                ls_before_SerialNo.append(a)
                ls_after_SerialNo.append(b)
```

```python
for col in ls:
    i = 0
    j = 0
    for idx, val in enumerate(df[col].values):
        if pd.isna(val):
            if col == 'Serial No.1':
                df.at[idx, col] = ls_after_SerialNo[i]
                i = i + 1
            elif col == 'Roll No':
                df.at[idx, col] = ls_after_roll[j]
                j = j + 1  
            
```

```python

regex = re.compile('\n')

for col in ls:
    i = 0
    j = 0
    for idx, val in enumerate(df[col].values):
        if pd.notna(val):  # Check for non-NaN values before applying the regular expression
            val_str = str(val)  # Convert the value to a string explicitly
            if regex.search(val_str):
                if col == 'Serial No.1':
                    df.at[idx, col] = ls_before_SerialNo[i]
                    i = i + 1
                elif col == 'Roll No':
                    df.at[idx, col] = ls_before_roll[j]
                    j = j + 1
```

```python
regex = re.compile('\n')
for idx , val in enumerate(df['Subject chosen'].values):
    val_str = str(val)
    if regex.search(val_str):
        x , y = val_str.split('\n' , 1) # Split only once, in case there are more than one newline character
        df.at[idx , 'Subject chosen'] = x.strip()
        
```
