```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
flights = sns.load_dataset('flights')
flights
```

```python
iris = sns.load_dataset('iris')
iris
```

# 1.Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(x='total_bill', y='tip',
                hue = tips.sex ,
                size = tips.size , style = tips.smoker,
                data=tips)
plt.show()
```

# 2. Bar Plot (Numerical - Categorical)

```python
titanic = pd.read_csv("C:\\Users\\arups\\OneDrive\\Desktop\\COLLECTION\\PROGRAMS\\TEST_CSVs\\train.csv")
titanic
```

```python
sns.barplot(x = titanic['Pclass'] , y = titanic['Fare'] , data = titanic)
plt.show() # IN TITANIC DATA SET 'Pclass' COLUMN CONSIDER AS A CATEGORICAL COLUMN
```

# 3. Box Plot (Numerical - Categorical)

```python
sns.boxplot(x = titanic['Sex'] , y = titanic['Age'] , hue = titanic.Survived , data = titanic)
plt.show()
```

# 4. Distplot(Numerical - Categorical)

```python
sns.distplot(titanic[titanic['Survived'] == 0]['Age'] , hist = False)
sns.distplot(titanic[titanic['Survived'] == 1]['Age'] , hist = False)
plt.show()
```

# 5. HeatMap (Categorical - Categorical)

```python
titanic.head(3)
```

```python
pd.crosstab(titanic['Pclass'] , titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'] , titanic['Survived']))
plt.show()
```

```python
survival_percentage = titanic.groupby('Pclass')['Survived'].mean() * 100
print(survival_percentage)

survival_percentage_sexwise = titanic.groupby(['Pclass' , 'Sex'])['Survived'].mean()*100
print(survival_percentage_sexwise)
```

# 6. ClusterMap (Categorical - Categorical)

```python
sns.clustermap(pd.crosstab(titanic['SibSp'] , titanic['Survived']))
plt.show()
```
