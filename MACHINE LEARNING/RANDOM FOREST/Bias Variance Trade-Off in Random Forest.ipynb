```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
np.random.seed(42)
X, y = make_circles(n_samples=500, factor=0.1, noise=0.35, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
def analyzer(max_depth, X, y):
    a = np.arange(X[:, 0].min() - 1, X[:, 0].max() + 1, step=0.01)
    b = np.arange(X[:, 1].min() - 1, X[:, 1].max() + 1, step=0.01)
    XX, YY = np.meshgrid(a, b)
    clf = DecisionTreeClassifier(max_depth=max_depth, random_state=42)
    clf.fit(X, y)
    input_array = np.array([XX.ravel(), YY.ravel()]).T
    Z = clf.predict(input_array)
    Z = Z.reshape(XX.shape)
    
    plt.contourf(XX, YY, Z, alpha=0.5)
    plt.scatter(X[:, 0], X[:, 1], c=y)
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.title(f"Decision Boundary (Max Depth = {max_depth})")
    print('Important features between X and y is {}'.format(clf.feature_importances_)) # Sum of all feature importance is 1
    plt.show()
```

```python
X, y = make_circles(n_samples=500, factor=0.1, noise=0.35, random_state=42)
analyzer(max_depth=10 , X=X , y=y)
```

```python
def analyzer_randomforest(n_estimators, X, y):
    a = np.arange(X[:, 0].min() - 1, X[:, 0].max() + 1, step=0.01)
    b = np.arange(X[:, 1].min() - 1, X[:, 1].max() + 1, step=0.01)
    XX, YY = np.meshgrid(a, b)
    clf = RandomForestClassifier(n_estimators=n_estimators, random_state=42)
    clf.fit(X, y)
    input_array = np.array([XX.ravel(), YY.ravel()]).T
    Z = clf.predict(input_array)
    Z = Z.reshape(XX.shape)
    
    plt.contourf(XX, YY, Z, alpha=0.5)
    plt.scatter(X[:, 0], X[:, 1], c=y)
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.title(f"Random Forest (Number Of Estimators= {n_estimators})")
    plt.show()
```

```python
X, y = make_circles(n_samples=500, factor=0.1, noise=0.35, random_state=42)
analyzer_randomforest(n_estimators=500 , X=X , y=y)
```

```python
# Generating Regression Data
n_train = 150        
n_test = 1000       
noise = 0.1
def f(x):
    x = x.ravel()
    return np.exp(-x ** 2) + 1.5 * np.exp(-(x - 2) ** 2)

def generate(n_samples, noise):
    X = np.random.rand(n_samples) * 10 - 5
    X = np.sort(X).ravel()
    y = np.exp(-X ** 2) + 1.5 * np.exp(-(X - 2) ** 2)\
        + np.random.normal(0.0, noise, n_samples)
    X = X.reshape((n_samples, 1))

    return X, y

X_train, y_train = generate(n_samples=n_train, noise=noise)
X_test, y_test = generate(n_samples=n_test, noise=noise)
```

```python
plt.figure(figsize=(18, 8))
plt.plot(X_test, f(X_test), "r")
plt.scatter(X_train, y_train, c="b", s=20)
plt.xlim([-5, 5])
     
```

```python
# One decision tree regressor
from sklearn.tree import DecisionTreeRegressor

dtree = DecisionTreeRegressor().fit(X_train, y_train)
d_predict = dtree.predict(X_test)

plt.figure(figsize=(18, 8))
plt.plot(X_test, f(X_test), "r")
plt.scatter(X_train, y_train, c="b", s=20)
plt.plot(X_test, d_predict, "g", lw=2)
plt.xlim([-5, 5])
plt.title("Decision tree, MSE = %.2f"%np.sum((y_test - d_predict) ** 2))
```

```python
# Random Forest
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor(n_estimators=1000).fit(X_train, y_train)
rf_predict = rfr.predict(X_test)

plt.figure(figsize=(18, 8))
plt.plot(X_test, f(X_test), "r")
plt.scatter(X_train, y_train, c="b", s=20)
plt.plot(X_test, rf_predict, "g", lw=2)
plt.xlim([-5, 5])
plt.title("Random forest, MSE = %.2f" % np.sum((y_test - rf_predict) ** 2));
```
