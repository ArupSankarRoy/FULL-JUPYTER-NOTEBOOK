```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
```

```python
def analyzer(max_depth):
    
        data = pd.read_csv("C:\\Users\\arups\\OneDrive\\Desktop\\COLLECTION\\PROGRAMS\\TEST_CSVs\\Social_Network_Ads.csv")
        X = data.iloc[:, 2:4]
        y = data.iloc[:, -1]

        clf = DecisionTreeClassifier(max_depth=max_depth)
        clf.fit(X, y)

        a = np.arange(start=X.iloc[:, 0].min() - 1, stop=X.iloc[:, 0].max() + 1, step=0.1)
        b = np.arange(start=X.iloc[:, 1].min() - 1, stop=X.iloc[:, 1].max() + 1, step=100)

        XX, YY = np.meshgrid(a, b)
        input_array = np.array([XX.ravel(), YY.ravel()]).T # With out transpose input_array has 594440 features, but DecisionTreeClassifier is expecting 2 features as input because of 'X'
        
        # OR :: input_set = np.c_[XX.ravel(), YY.ravel()]
        
        labels = clf.predict(input_array) # NOTE : As input X has 2 feature predict() also want it should have 2 feature because it deals with new data sets and there no of feature 
                                          # Must be match with input data
        
        # input_array is considered as new test data because the decision tree classifier has not seen this data before.
        # The decision tree classifier is used to predict the labels for the data points in the input_array array based on the training data that it was trained on.

        plt.contourf(XX, YY, labels.reshape(XX.shape), alpha=0.5) # the regions in a decision tree are determined by the predictions made by the predict() method and
                                                                  # Here labels.reshape(YY.shape) is represent Z axia and contourf() use for representiog Z value and it must be 2D, not 1D
        plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c= y)
        plt.show()
```

```python
analyzer(max_depth = 3)
```
