## Linear Transformations rotate and scale most vectors but only scale Eigen vectors

```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 8,8
```

# (9 4
#  4 3)

```python
x = np.linspace(-1 , 1 , 100)
```

```python
y1 = np.sqrt(1 - np.square(x))
y2 = -1*y1
```

```python
plt.plot(x , y1 , color = 'b')
plt.plot(x , y2 , color = 'b')
plt.gca().set_aspect('equal') 
# plt.gca(): gca stands for "get current axis." It returns the current Axes instance being used by the plot.
#  By passing 'equal' as the first argument, it instructs the plot to keep the aspect ratio equal, which means one unit on the x-axis will have the same physical length as one unit on the y-axis.                           
plt.xlim([-2, 2])
plt.ylim([-2, 2])
plt.show()
```

```python
def transformation(x , y):
    return 9*x + 4*y , 4*x + 3*y
```

```python
x_new1, y_new1 = transformation(x,y1)
x_new2, y_new2 = transformation(x,y2)
plt.plot(x_new1,y_new1, 'r')
plt.plot(x_new2,y_new2, 'r')
plt.show()
```

```python
eig_vals, eig_vecs = np.linalg.eig(np.array([[9,4],[4,3]]))
print('Eigenvectors \n%s' %eig_vecs)
print('\nEigenvalues \n%s' %eig_vals)
```

```python
soa = np.array([[0, 0, 
                 eig_vals[0] * eig_vecs[0][0], 
                 eig_vals[0] * eig_vecs[1][0]]])

soa1 = np.array([[0, 0, 
                  eig_vals[1] * eig_vecs[0][1], 
                  eig_vals[1] * eig_vecs[1][1]]])

X, Y, U, V = zip(*soa)
X1, Y1, U1, V1 = zip(*soa1)

plt.plot(x,y1, 'b')
plt.plot(x,y2, 'b')
plt.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1)
plt.quiver(X1, Y1, U1, V1, angles='xy', scale_units='xy', scale=1)
plt.xlim([-1, 10])
plt.ylim([-1, 10])
plt.show()
```

```python
print('soa is {} '.format(soa))

X, Y, U, V = zip(*soa)
print('x is {} y is {} u is {} v is {} '.format(X,Y,U,V))
```

```python
soa = np.array([[0, 0, 
                 eig_vals[0] * eig_vecs[0][0], 
                 eig_vals[0] * eig_vecs[1][0]]])

soa1 = np.array([[0, 0, 
                  eig_vals[1] * eig_vecs[0][1], 
                  eig_vals[1] * eig_vecs[1][1]]])

X, Y, U, V = zip(*soa)
X1, Y1, U1, V1 = zip(*soa1)
plt.plot(x_new1,y_new1, 'r')
plt.plot(x_new2,y_new2, 'r')
plt.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1)
plt.quiver(X1, Y1, U1, V1, angles='xy', scale_units='xy', scale=1)
plt.draw()
plt.show()
```
