```python
# The .fit_transform() method in scikit-learn expects a 2D array as input because
# it is designed to work with tabular data where each column represents a feature,
# and each row represents a sample or observation.

from sklearn.preprocessing import OneHotEncoder
import numpy as np

categories = np.array(['A', 'B', 'A', 'C', 'B', 'A'])

# Initialize OneHotEncoder with sparse_output=True
oe = OneHotEncoder(sparse_output=True)

# Perform one-hot encoding
encoded_data = oe.fit_transform(categories.reshape(-1 , 1))

#-------------------------------------------------------------------------------------------------------------------------------------------------------------
# In the context of categories.reshape(-1, 1), the -1 indicates that
# you want to reshape the categories array into a 2D array with an unknown number of rows but with a single column.

# For example, if categories has 6 elements (i.e., categories.size is 6), then categories.reshape(-1, 1) will result in a 2D array with 6 rows and 1 column.
# If categories has 10 elements, the resulting array will have 10 rows and 1 column. The -1 acts as a placeholder for the inferred dimension.
#--------------------------------------------------------------------------------------------------------------------------------------------------------------
print(encoded_data)
print(categories.reshape(-1, 1))
print(np.size(categories))
```
