```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random
```

```python
na , nb = (5 , 3)

a = np.linspace(1 , 2 , na)
b = np.linspace(1 , 2 , nb)

XX , YY = np.meshgrid(a , b)
print(XX)
print(YY)
```

```python
na , nb = (5 , 3)

a = np.linspace(1 , 2 , na)
b = np.linspace(1 , 2 , nb)

XX , YY = np.meshgrid(a , b , sparse=True)
print(XX)
print(YY)
```

```python
x_axis = np.linspace(-4 , 4 , 9)
y_axis = np.linspace(-5 , 5 , 11)
XX , YY = np.meshgrid(x_axis , y_axis)
eq = XX **2 + 4 * YY **2 
plt.contourf(x_axis , y_axis , eq , cmap = 'jet')
plt.colorbar()
plt.show()
```

```python
random_data = np.random.random((11 , 9))
print(x_axis.shape)
print(y_axis.shape)

plt.contourf(x_axis , y_axis , random_data , cmap = 'jet')
plt.show()
```
