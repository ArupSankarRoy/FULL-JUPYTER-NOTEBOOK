```python
import plotly.offline as pyo
import pandas as pd
import matplotlib.pyplot as plt
```

```python
orders = pd.read_excel("sales.xlsx")
orders
```

```python
plt.plot(orders.Sales)
plt.show()
```

```python
pyo.plot([{
    
    "x": orders.index,
    "y": orders.Sales
}])
```

```python
orders_sf = orders[["Sales" , "Profit"]]
```

```python
orders_sf.head()
```

```python
pyo.plot([{
    
    "x": orders_sf.index,
    "y": orders_sf[col],
    "name" : col         # If u don't put name:col then as a legend it shows,
                         # trace 0 ,trace 1
} for col in orders_sf.columns])

plt.show()
```
