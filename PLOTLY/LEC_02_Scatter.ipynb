# PLOTLY SCATTER CHART

```python
import plotly.offline as pyo
import pandas as pd
import matplotlib.pyplot as plt
```

```python
orders = pd.read_excel("sales.xlsx")
orders
```

```python
import plotly.graph_objs as go
```

```python
orders_sf = orders[["Sales" , "Profit"]].head()
orders_sf
```

```python
import plotly.graph_objs as go
import plotly.offline as pyo

data = go.Scatter(
    x=orders_sf.Sales,
    y=orders_sf.Profit,
    mode='markers',
    marker = dict(
        
        size = 14,
        color = 'rgb(120 , 150 , 180)',
        symbol = "square",
        line = dict(        # It basically the Border Width of each Block.
            width = 2,
        
        )
    )
)

layout = go.Layout(
    title=dict(text='Sales vs. Profit', font=dict(color='red', family='Arial')),
    xaxis=dict(title='Sales', tickfont=dict(color='red', family='Arial')),
    yaxis=dict(title='Profit', tickfont=dict(color='red', family='Arial')),
)

fig = go.Figure(data=[data], layout=layout)
pyo.plot(fig)
plt.show()
```
