```python
whether_data = {
    
    "day" : ["1/01/2017" , "1/02/2017" , "1/03/2017"],
    "temperature" : [32 , 35 , 38],
    "windspeed" : [6 , 7 , 2],
    "event" : ["Rain" , "Sunny" , "Snow"]
}

df = pd.DataFrame(whether_data)
df
```

```python
whether_data = [
    
    ("1/01/2017" , 32 , 6 , "Rain"),  # Row wise
    ("1/02/2017" , 35 , 7 , "Sunny"),
    ("1/03/2017" , 38 , 2 , "Snow")

]

df = pd.DataFrame(whether_data , columns = ["day" , "temperature" , "windspeed" , "event"])
df
```

```python
weather_data = [
    {'day': '1/1/2017', 'temperature': 32, 'windspeed': 6, 'event': 'Rain'},  # Each element of List is Dictionary that contains Column-value pair.
    {'day': '1/2/2017', 'temperature': 35, 'windspeed': 7, 'event': 'Sunny'},
    {'day': '1/3/2017', 'temperature': 28, 'windspeed': 2, 'event': 'Snow'},
    
]
df = pd.DataFrame(data=weather_data, columns=['day','temperature','windspeed','event'])
df
```
