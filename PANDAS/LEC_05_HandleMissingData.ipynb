```python
import pandas as pd

df = pd.read_csv("C:\\Users\\Asus\\Desktop\\Jupyter Notebook\\PANDAS_02\\py-master\\py-master\\pandas\\5_handling_missing_data_fillna_dropna_interpolate\\weather_data.csv" , parse_dates = ["day"])

df.set_index("day" , inplace = True)
df
```

```python
new_df = df.fillna(method = "ffill" , limit = 1)
# This will fill replace the "NaN" value only 1 time, 
new_df
```

```python
new_df = df.interpolate() #  means blank data ager and pi6er datar ka6aka6i thakbe
new_df
```

###     IN ABOVE CASE,
###                   2017-01-01 -> 32.000000
###                   2017-01-04 -> 30.000000
###                   2017-01-05 -> 28.000000
###     Look 30.00000 temperature is between the 32.000000     and 28.000000
###     and we skip  2nd and 3rd jan. So doesn't the temperature of date 017-01-04
###     is more close to 017-01-05 .
###     To do that we use method = "time"

```python

new_df = df.interpolate(method = "time") 
new_df
```
