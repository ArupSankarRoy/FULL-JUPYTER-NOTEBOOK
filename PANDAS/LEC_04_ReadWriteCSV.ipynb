## CONVERTERS:

```python
import pandas as pd

def convert_people_cell(cell):
    if cell == "n.a." or cell =="not available":
        return "Sam Walton"
    return cell

def convert_eps_cell(cell):
     if cell == "n.a." or cell =="not available":
        return None
     return cell

df = pd.read_excel("C:\\Users\\Asus\\Desktop\\Jupyter Notebook\\PANDAS_02\\py-master\\py-master\\pandas\\4_read_write_to_excel\\stock_data.xlsx",
                  converters= {
                      "people" : convert_people_cell,
                       "eps" : convert_eps_cell
                  })
df
```
